Import Library

```python
from random import randint
from time import time
```

Bubble Sort

```python
# Bubble Sort 

# Tanpa Optimasi
def bubbleSort(arr):
    n = len(arr)
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]


```

Merge Sort

```python
#mergesort 
def mergeSort(a):
    width = 1    
    n = len(a)
    while (width < n):
        l=0;
        while (l < n): 
            r = min(l+(width*2-1), n-1)         
            m = min(l+width-1,n-1)        
            merge(a, l, m, r)
            l += width*2
        width *= 2
    return a
    
def merge(a, l, m, r): 
    n1 = m - l + 1
    n2 = r - m 
    L = [0] * n1 
    R = [0] * n2 
    for i in range(0, n1): 
        L[i] = a[l + i] 
    for i in range(0, n2): 
        R[i] = a[m + i + 1] 
    i, j, k = 0, 0, l 
    while i < n1 and j < n2: 
        if L[i] <= R[j]: 
            a[k] = L[i] 
            i += 1
        else: 
            a[k] = R[j] 
            j += 1
        k += 1
    while i < n1: 
        a[k] = L[i] 
        i += 1
        k += 1
    while j < n2: 
        a[k] = R[j] 
        j += 1
        k += 1
```

```python
timeBubble = []
timeMerge = []
```

Data Set 3000

```python
# Array Belum Terurut

def generate_randomArray(n):
  a = []
  for i in range(n):
    a.append(randint(1,10000))
  return a

arr1 = generate_randomArray(3000)
arr2 = arr1[:]

print("Array Awal  :", arr1)

# Bubble Sort
bubbleSort(arr1)
print("Bubble Sort :", arr1)

# Merge Sort
mergeSort(arr2)
print("Merge Sort  :", arr2)
```

```python
#Menghitung waktu eksekusi

start1 = time()
bubbleSort(arr1)
end1 = time()-start1

start2 = time()
mergeSort(arr2)
end2 = time()-start2

print("Bubble Sort ends in " , (end1), " seconds. ")
print("Merge Sort ends in " , (end2) , " seconds. ")

timeBubble.append(end1)
timeMerge.append(end2)
```
